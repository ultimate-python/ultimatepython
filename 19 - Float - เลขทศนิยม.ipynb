## 1) `Float` คือ ข้อมูลประเภททศนิยม

`float` เป็นประเภทของ `object` ที่เน้นเก็บข้อมูลที่มีคุณสมบัติเป็น **ทศนิยม**

## 2) สร้าง `integer` โดยการพิมพ์ `ตัวเลข` แล้วตามด้วย `.`

เพื่อพิมพ์ตัวเลขและตามด้วย `.` จะเป็นการสร้าง `object` ประเภท `float` ทันที

```python
10.
```

## 3) `float` ปรับการแสดงผลทศนิยมได้

`integer` และ `float` มีค่าทางคณิตศาสตร์เท่ากัน ต่างกันที่ `float` ปรับการแสดงผลทศนิยมได้

## 4) `float formatting` ปรับการแสดงผลทศนิยม

```python
weight = 50.123456789
```

ส่วนของการเปลี่ยนแปลงข้อมูล 
`{` `ตัวแปร` `:` `{จำนวนตัวอักษรขั้นต่ำที่แสดง รวมจุด}` `.` `{จำนวนตัวเลขที่แสดง}` `}`

```python
print(f"His weight is {weight:{5}.{5}}")
```
