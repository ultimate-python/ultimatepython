## 2) `import` `docx` มาใช้บน Jupyter Notebook

นำ `Document` จาก `docx` มาใช้

```python
from docx import Document
```

## 3) สร้าง/เปิดไฟล์ Words ด้วย `Document()`

`Document()` สร้าง `object` ที่เก็บไฟล์ Words โดยการเปิดไฟล์ที่มีอยู่แล้ว หรือสร้างไฟล์ใหม่ 

```python
doc = Document('invoice.docx')
```

## 4) อ่านตารางใน Words ด้วย `.tables[]`

`.tables[]` เลือก table ใน Words มาทำงานด้วย

```python
table1 = doc.tables[0]
```

## 5) เลือก `cell` ที่จะทำงานด้วย  `.cell`

`.cell()` รับค่า `index` ของ `row` และ `index` ของ `column`

เปลี่ยนเนื้อหาด้วย `.text`

```python
table1.cell(0,0).text
```

```python
table1.cell(0,1).text
```

```python
table2 = doc.tables[1]
```

```python
table2.cell(1,0).text
```

```python
table2.cell(1,1).text
```

```python
table2.cell(1,2).text
```

```python
table2.cell(1,3).text
```

## 5) เลือก `paragraph` ที่จะทำงานด้วย  `.paragraphs[]`

เลือก `paragraph` ที่จะทำงานด้วย `.paragraphs[]` และทำงานกับข้อความด้วย`.text`

```python
doc.paragraphs[5].text
```

```python
doc.paragraphs[6].text
```

```python
doc.paragraphs[7].text
```

## 6) บันทึก Words ด้วย  `.save()`

ใช้ `.save()` ที่รับค่าชื่อไฟล์ที่ต้องการเป็น `string` ที่มีชื่อสกุลไฟล์ด้วย

```python
doc.save('invoice2.docx')
```
