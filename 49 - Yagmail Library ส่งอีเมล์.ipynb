## 7) แนบรูปด้วย `.inline()`

```python
body_pic = yagmail.inline("if.png")
```

## 8) ส่งอีเมล์ด้วย `.send()`

แนบรูปเข้าไปในข้อความในจดหมาย

รวบรวมเนื้อหาจดหมายใน list

```python
email_body = [body,body_pic]
```

รวบรวมไฟล์ที่จะแนบไปกับอีเมล์เป็น list

ส่ง email โดยใช้รายละเอียดดังกล่าว

```python
yag.send(
    to=receiver,
    subject=subject,
    contents=[body,body_pic], 
    attachments="if.png",
)
```
