## 1) `integer` คือ ข้อมูลประเภทจำนวนเต็ม

`string` เป็นประเภทของ `object` ที่เน้นเก็บข้อมูลที่มีคุณสมบัติเป็น **ตัวเลข จำนวนเต็ม**

## 2) สร้าง `integer` โดยการพิมพ์ `ตัวเลข`

เพื่อพิมพ์ตัวเลขจะมี default เป็น `object` ประเภท `integer` ทันที

```python
10
```

## 3) `integer` มีค่าทางคณิตศาสตร์

```python
10 + 10
```

ตัวอย่าง

1) `+` การบวก

2) `-` การลบ

3) `*` การคูณ

4) `/` การหาร

5) `**` การยกกำลัง

6) `//` การหารปัดเศษทิ้ง

7) `%` การหารหาเศษ

```python
19//10
```
