# Concepts สำคัญ

การทำความเข้าใจภาษาคอมพิวเตอร์เป็นเรื่องที่อาศัยมุมมองที่ถูกต้อง เพราะหลายครั้งเป็นเรื่องใหม่ที่ไม่เคยใช้ในชีวิตประจำวัน ทำให้เกิดความไม่คุ้นชิน และทำให้การเรียนรู้เป็นไปได้ช้า ดังนั้นก่อนที่จะเริ่มเรียนเรามาทำความเข้าใจคอนเซปสำคัญเหล่านี้ โดยการเปรียบเทียบกับสิ่งที่เราคุ้นเคย เพื่อให้เห็นภาพใหญ่ของคอนเซปเหล่านี้ และให้ทำความเข้าใจได้ง่ายมากขึ้น

## 1) Python คือ ภาษา

ผมอยากให้ทุกคนมองว่า Python เป็นภาษาหนึ่ง แต่ในขณะที่ภาษาปกติที่เราใช้สื่อสารกันเน้นการสื่อสารระหว่างบุคคล ภาษา Python เป็นภาษาที่ใช้สื่อสารระหว่างคน และคอมพิวเตอร์ ดังนั้นการเรียน Python ก็คือการเรียนภาษาที่เราจะต้องเข้าใจโครงสร้างภาษา การผูกประโยคหรือคำสั่ง ให้คอมพิวเตอร์ทำตามขั้นตอนจนกระทั่งเราได้สิ่งที่เราต้องการ

**เทคนิค:** เมื่อต้องการเขียนคำสั่งใดๆ เป็นภาษา Python ลองสั่งคำสั่งนั้นให้ชัดเจนเป็นคำพูด หรือการเขียนก่อน แล้วค่อยแปลงเป็นภาษา Python

## 2) Object Oriented Programming Langauge

ข้อมูลในภาษา Python จะถูกเรียกว่า Object หรือวัตถุ หรือ item ซึ่งเมื่อสร้างข้อมูลใดๆ ขึ้นมาแล้วจะต้องกำหนดประเภทของข้อมูลขึ้นมากำกับด้วย ซึ่งการกำหนดประเภทข้อมูล และชนิดข้อมูลจะเป็นสิ่งที่เราเรียนรู้กันในบทถัดๆ ไป

Python มีตัวช่วยที่จะทำให้เรากำหนดประเภทข้อมูลหรือจำแนกประเภทข้อมูลได้ง่ายด้วยการใช้สัญลักษณ์เพื่อกำหนดประเภทข้อมูล และการแบ่งข้อมูลแต่ละประเภทออกจากกันด้วยสี

**ข้อสังเกต:** Python จำแนกข้อมูลประเภทต่างๆ ออกจากกันด้วยสี

เป็นคุณสมบัติของ Python ที่กำหนดให้ทุกข้อมูลจะต้องถูกสร้างพร้อมกับเลือกประเภทข้อมูล และการทำงานใดๆ กับข้อมูลจะถูกจำกัดด้วยประเภทข้อมูลของข้อมูลนั้นๆ และกำหนดข้อมูลเป็นประเภทใดๆ แล้วเราจะสามารถเข้าถึงเครื่องมือสำเร็จรูปที่ใช้เพื่อจัดการข้อมูลประเภทนั้นๆ ได้ทันทีทำให้การจัดการข้อมูลเป็นไปได้ง่ายมากขึ้น ซึ่งจะได้เรียนรู้เพิ่มเติมเมื่อเรียนบทต่างๆ เกี่ยวกับประเภทของข้อมูล
