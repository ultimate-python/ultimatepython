## 1) `method` - คำสั่งเฉพาะประเภทข้อมูล

`method` เป็น `function` ที่ถูกสร้างไว้พร้อมกับประเภทข้อมูลทำให้จัดการข้อมูลได้ง่ายมากขึ้น

## 2) ใช้ `method` ด้วยสัญลักษณ์ `.`

เมื่อสร้าง `object` แล้วให้พิมพ์ `.` ตามด้วย `method` ที่ต้องการใช้ และใส่ `( )` เพื่อรับ `input` เพราะ `method` เองก็เป็น `function`

## 3) `string.split( )` - แยกคำออกจากกัน

ใช้เพื่อแยกคำออกจากกันที่ตำแหน่ง `space bar` ที่จะได้ `output` เป็น `object` ประเภท `list`

```python
'hello world'.split()
```

## 4) `string.startwith( )` - เช็คตัวอักษรเริ่มต้น

ใช้เช็คว่า `object` ตัวแรกเริ่มต้นด้วยตัวอักษรที่เป็น `input` หรือไม่ โดยได้ `output` เป็น `object` ประเภท `boolean`

```python
'hello world'.startswith('h')
```

## 5) `tab completion` - กด tab ดูเครื่องมือทั้งหมด

`tab completion` คือ การใช้ประโยชน์จากโครงสร้างของ python ที่จะโชว์เครื่องมือทั้งหมด เมื่อกดปุ่ม `tab` หลังจากที่เราพิมพ์ `.` ไปด้านหลัง `object` หรือในกรณีอื่นๆ

```python
'hello world'
```
