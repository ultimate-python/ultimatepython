## 1) ปรับขนาดตัวอักษร

การเขียนหัวข้อให้เกิดเป็นตัวอักษรตัวใหญ่ทำได้ 6 ระดับ จากการใส่สัญลักษณ์ # ลงไปใน Markdown cell ตามตัวอย่างด้านล่าง

# hello

## hello

### hello

#### hello

##### hello

###### hello

## 3) การเขียนโน้ตลงใน code cell

นอกจากเราจะเขียนเอกสารลงบน markdown cell ได้แล้วยังสามารถเขียนโน้ตลงบน code cell ได้ด้วย โดยใช้ สัญลักษณ์ #

```python
#การเขียนโน้ตโดยใช้ # ทุกตัวอักษรที่อยู่หลัง # ตัวแรกถูกจัดให้เป็นโน้ตทั้งหมด
```

จะสังเกตได้ว่า # จะเปลี่ยนตัวอักษรด้านหลัง # ในบรรทัดนั้นให้กลายเป็นโน้ตที่จะไม่ถูกประมวลผลเป็นภาษา Python

## 4) การสร้าง list

การสร้าง list สามารถทำได้ใน markdown cell ด้วยการเว้นวรรคด้านของข้อมูล 1 tab ซึ่งสามารถใส่ tab ลงไปได้ด้วยการกด `tab`

#### ตัวอย่าง markdown cell ที่มี list อยู่

หัวข้อ

    1.
    2.
    3.
