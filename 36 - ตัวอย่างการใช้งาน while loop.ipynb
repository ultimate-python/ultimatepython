## 1) ใช้การนับรอบเป็นเงื่อนไข

```python
count = 0
while count < 5:
    print('The statement is true!')
    count += 1
```

## 2) ใช้การเปลี่ยนค่า `boolean` เป็นเงื่อนไข

```python
run = True
count = 0
while run:
    if count == 5:
        run = False
    print('The statement is true!')
    count += 1
```

## 3) ใช้ `break` หยุด

```python
count = 0
while True:
    if count == 5:
        break
    print('The statement is true!')
    count += 1
```
