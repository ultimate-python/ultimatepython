```python
count = 0
while count < 5:
    print('The statement is true!')
    count += 1
```

## 2)  `boolean` 

```python
run = True
count = 0
while run:
    if count == 5:
        run = False
    print('The statement is true!')
    count += 1
```

## 3)  `break` 

```python
count = 0
while True:
    if count == 5:
        break
    print('The statement is true!')
    count += 1
```
