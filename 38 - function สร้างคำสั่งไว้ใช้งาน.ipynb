## 1) สร้าง `function` ด้วย `def`

เราสามารถจัดเก็บการกระทำไว้ในรูปแบบของ `function` ด้วยการสร้างผ่าน `def`

```python
def hello():
    print('hello')
```

```python
hello()
```

## 2) ให้ `function` รับ `input` ผ่าน `()`

เราสามารถกำหนด `input` ผ่านการระบุ `ตัวแปร` ใน `()`

```python
def hello(name):
    print(f'hello {name}')
```

```python
hello('john')
```

## 3) ให้ `function` สร้าง `object` ผ่าน `return`

สร้าง `object` ผ่าน `return`

```python
def hello(name):
    return 'hello' + ' ' + name
```

```python
hello('john')
```
