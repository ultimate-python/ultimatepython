## 1)  `function`  `def`

 `function`  `def`

```python
def hello():
    print('hello')
```

```python
hello()
```

## 2)  `function`  `input`  `()`

 `input`  ``  `()`

```python
def hello(name):
    print(f'hello {name}')
```

```python
hello('john')
```

## 3)  `function`  `object`  `return`

 `object`  `return`

```python
def hello(name):
    return 'hello' + ' ' + name
```

```python
hello('john')
```
