## 2) `import` `pptx` มาใช้บน Jupyter Notebook

นำ `Presentation` จาก `pptx` มาใช้

```python
from pptx import Presentation
```

## 3) สร้าง PowerPoint ด้วย `Presentation()`

`Presentation()` สร้าง `object` ที่เก็บไฟล์ PowerPoint 

```python
prs = Presentation()
```

## 4) เลือกรูปแบบของ slide ด้วย `.slide_layouts[]`

เลือกรูปแบบ slide จาก `.slide_layouts[]`

```python
slide3_register = prs.slide_layouts[8]
```

## 5) เพิ่ม slide ลงไปใน PowerPoint ด้วย `.slides.add_slide()`

เพิ่ม slide ด้วย `.slides.add_slide()` และเก็บ slide ที่สร้างไปปรับแต่ง

```python
slide3 = prs.slides.add_slide(slide3_register)
```

## 6) เปลี่ยนหัวของ slide ด้วย `.shapes.titles.text`

เพิ่มหัวข้อลงใน slide ด้วย `.shapes.title.text`

```python
slide3.shapes.title.text = 'Title'
```

## 7) ใส่ข้อความลงใน placeholder ด้วย  `.placeholders[].text`

เลือก placeholders ที่ต้องการด้วย `placeholders[]` และเขียน text ลงไปด้วย `placeholders[].text`

```python
slide3.placeholders[2].text = 'Subtitle1'
```

## 8) ใส่รูปลงไปใน placeholder ด้วย  `.placeholders[].insert_picture()`

เลือก placeholders ที่ต้องการด้วย `placeholders[]` และแนบรูปลงไปด้วย `placeholders[].insert_picture()`

```python
slide3.placeholders[1].insert_picture('list.png')
```

การใส่รูปลงไปจะถูกตัดให้เท่ากับอัตราส่วนของ placeholder

```python
slide3.placeholders[1].height
```

```python
slide3.placeholders[1].widthth
```

## 9) บันทึก PowerPoint ด้วย  `.save()`

ใช้ `.save()` ที่รับค่าชื่อไฟล์ที่ต้องการเป็น `string` ที่มีชื่อสกุลไฟล์ด้วย

```python
prs.save('test99.pptx')
```
