## 1) `seires` ชุดข้อมูลที่เก็บด้วย `index`

ชุดข้อมูลบางประเภทต้องใช้การระบุข้อมูลที่เฉพาะเจาะจง เช่น time series ที่ต้องใช้ เวลา เป็นการระบุ ข้อมูลพวกนี้จะถูกเก็บใน `object` ประเภท `series` ที่สามารถเปลี่ยน `index` เป็นค่าอะไรก็ได้

ข้อมูลของ `series` ประกอบด้วย 2 ส่วน คือ `data` และ `index`

## 2) `import` `pandas` นำ `pandas` มาใช้บน Jupyter Notebook

`series` เป็น `object` ที่ถูกสร้างขึ้นด้วย `library` `pandas` ดังนั้นจะใช้งานได้จะต้องเรียก `pandas` มาไว้บน Jupyter Notebook ก่อน

```python
import pandas as pd
```

## 2) สร้าง `seires` ผ่าน `pd.Series()`

สร้าง `series` เปล่า

```python
pd.Series()
```

## 3) กำหนด `data` ด้วย `list` `object`

`data` ที่จะเป็นข้อมูลของ `series` นิยมเริ่มด้วย `list` `object`

```python
data = [1,2,3]
```

## 4) กำหนด `index` ด้วย `list` `object`

`index` ที่จะเป็นข้อมูลของ `series` นิยมเริ่มด้วย `list` `object`

```python
index = ['a','b','c']
```

## 5) `seires` ประกอบด้วย `data` และ `index`

สิ่งที่ `series` เก็บประกอบด้วยข้อมูลหรือ `data` ที่ระบุด้วย `index`

```python
ser = pd.Series(data=data,index=index)
ser
```

## 6) ดึงข้อมูลด้วย `index`

```python
ser['a']
```

## 7) การทำงานกับ `seires`: `seires` + `seires`

```python
ser1 = pd.Series(data=[1,2,3,4],index=['a','b','c','d'])
ser1
```

```python
ser2 = pd.Series(data=[5,6,7,8],index=['a','b','c','e'])
ser2
```

```python
ser1 + ser2
```
