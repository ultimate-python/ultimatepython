## 1) `built-in function` - คำสั่งพื้นฐานที่เรียกใช้ได้ทันที

`built-in function` คือ คำสั่งที่เรียกใช้ได้ทันที เพราะถูกสร้างมาพร้อมกับตัว python ซึ่งส่วนใหญ่เป็นคำสั่งพื้นฐานง่ายๆ

ตัวอย่าง `built-in function`

```python
print
```

```python
len
```

```python
int
```

**ข้อสังเกต:** `built-in function` จะมีสีเขียว

#### `built-in function` ที่ใช้สร้าง `object` ประเภทต่างๆ

```python
str()
```

```python
int()
```

```python
float()
```

## 2) `input` `function` `output` concept

ในการสั่งคำสั่งจะต้องประกอบไปด้วย

1) `input` ข้อมูลที่ใส่เข้าไปเป็น `object`

2) `function` กระบวนการที่จัดไว้

3) `output` ข้อมูลที่ได้กลับมาเป็น `object`

## 3) `built-in function` รับ input ด้วย `( )`

การใช้งาน `built-in function` จะเริ่มด้วยชื่อคำสั่งตามด้วย `( )`

ตามลักษณะดังนี้ `ชื่อคำสั่ง(input)`

```python
print('hello world')
```

## 4) `print( )` คำสั่งให้แสดงผล

ใช้เพื่อการแสดงผลให้สวยงาม ไม่มี `output` ที่เป็น `object`

```python
print('hello world')
```

## 5) `len( )` คำสั่งให้นับ `object` ที่เก็บไว้

```python
len('hello world')
```
