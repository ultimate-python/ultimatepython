## 1) `if` เงื่อนไข "ทำเมื่อไร"

เงื่อนไข `if` ใช้เมื่อต้องการแยกการกระทำที่ไม่เหมือนกัน ออกเป็นกรณีต่างๆ ด้วยการกำหนดเงื่อนไข

`if` ใช้ร่วมกับ `boolean` และจะทำงานเมื่อรับค่า `True`

## 2) `if` ถ้าเกิด...ให้ทำ...

การเขียน `if` ให้นึกถึงเงื่อนไขที่บอกว่า ถ้าเกิด...ให้ทำ... 
โดยให้เขียนในลักษณะ `if` `เงื่อนไข` `:`

```python
a = 10
b = 20
```

```python
if a < b:
    print("a is less than b")
```

## 3) `elif` แต่ถ้าเกิด...ให้ทำ...

การเขียน `elif` เป็นส่วนขยายของ `if` ที่ทำให้สร้างเงื่อนไขได้ซับซ้อนมากขึ้นให้นึกถึงเงื่อนไขที่บอกว่า ถ้าเกิด...ให้ทำ... โดยให้เขียนในลักษณะ `elif` `เงื่อนไข` `:`

```python
a = 30
b = 20
```

```python
if a < b:
    print("a is less than b")
elif a > b:
    print("a is more than b")
```

## 4) `else` ถ้าไม่เกิดขึ้น ให้ทำ...

การเขียน `else` เป็นส่วนขยายของ `if` ที่บอกว่า ถ้าไม่เกิดขึ้น ให้ทำ... โดยให้เขียนในลักษณะ `else` `:`โดยไม่ต้องเขียนเงื่อนไขเพิ่มเติม

```python
a = 50
b = 50
```

```python
if a < b:
    print("a is less than b")
elif a > b:
    print("a is more than b")
else:
    print("a is equal to b")
```
