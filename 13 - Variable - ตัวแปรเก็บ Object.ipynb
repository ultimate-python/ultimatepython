## 1) Variable

หรือ ตัวแปร เป็นการตั้งชื่อให้กับ Object เมื่อพูดถึงตัวแปรผมอยากให้ทุกคนคิดภาพในหัวว่าตัวแปรเป็นถังน้ำที่มีชื่อแปะอยู่ และ Object ที่แทนด้วยตัวแปรนั้นๆ เป็นวัตถุชิ้นหนึ่งที่ถูกนำมาใส่ในถังนั้น

ตัวแปรใน Python สามารถตั้งเป็นชื่ออะไรก็ได้ สามารถเก็บ Object อะไรไว้ก็ได้ และสามารถเปลี่ยนไปเก็บ Object อื่นๆ เมื่อไรก็ได้ ทำให้มีความสะดวกในการเปลี่ยนแปลง แต่ก็ต้องคอยเช็คว่าตัวแปรนั้นๆ ยังเป็นตัวแทนของ Object ที่เราต้องการใช้งานไหม

**ตัวอย่าง**

```python
hello = 'hello world'
```

#### เรียกชื่อตัวแปร
