## 1) `for loop` เงื่อนไข "ทำกับทุก `object`"

เงื่อนไข `for loop` ใช้เมื่อต้องการสั่งให้การกระทำเกิดกับทุก `object` ใน `object` ที่กำหนด

## 2) `for loop` ทำงานกับ `iterable`

`iterable` คือ เป็นชื่อเรียกประเภทของ `object` ที่สามารถดึงข้อมูลออกมาที่ละ `object` ได้ 

ตัวอย่างข้อมูลที่ เป็น `iterable`

1.`string`
2.`list`
3.`Series`
4.`DataFrame`

[เรียนรู้เพิ่มเติม](https://www.pythonlikeyoumeanit.com/Module2_EssentialsOfPython/Iterables.html#:~:text=An%20iterable%20is%20any%20Python,over%20in%20a%20for%2Dloop)

## 3) การใช้งาน `for loop`

ใช้คำสั่ง และวิธีการเขียนดังตัวอย่างด้านล่าง

```python
name_list = ['somchai','somying','john']
```

```python
for name in name_list:
    print(name)
```
