## 1) Jupyter Notebook - ตัวเขียนโค้ดที่ใช้ง่ายที่สุด

ด้วยหน้าตา และการทำงานคล้ายกับ Words ที่คุ้นเคย และการแบ่งสัดส่วนเป็น `cell` ทำให้เข้าใจได้ง่าย

## 2) คลิกเลือก `cell`

ในการทำงานกับ `cell` ให้คลิกบริเวณที่ต้องการ `cell` จะถูกเลือก โดยมีสัญลักษณ์เป็นการตีกรอบ

## 3) การสร้าง `cell` ใหม่

เลือก`cell` กดปุ่ม `+` ในแถบเครื่องมือเพื่อสร้าง `cell` ด้านล่าง `cell` ที่กำลังเลือกอยู่

#### ปุ่มเพิ่ม cell บนแถบเครื่องมือ

## 4) การลบ `cell`

เลือกที่ `cell` ที่จะลบ แล้วเลือกเครื่องมือบนแถบเครื่องมือ `Edit` >> `Delete Cells`

#### ปุ่มลบ cell บนแถบเครื่องมือ

## 5) `code cell` - `cell` ที่ใช้เขียน python 

เมื่อสร้าง `cell` ขึ้นมา `cell` ที่จะได้ คือ `code cell` ที่ใช้เขียน python

## 6) เปลี่ยนชนิด `cell`

#### ปุ่มเปลี่ยนประเภท cell บนแถบเครื่องมือ

## 7) `markdown cell` - `cell` ที่ใช้เขียนบันทึก

เมื่อเปลี่ยนประเภทเป็น `markdown cell` จะได้ `cell` ที่ใช้จดบันทึก

## 8) `run` `cell` - สั่งให้ประมวลผล

เมื่อต้องการให้ `cell` ประมวลผล ให้ `run` `cell` นั้น

#### ปุ่ม Run ตามภาพด้านล่าง ด้านบนในแถบเครื่องมือ
