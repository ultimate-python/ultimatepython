## 1) นับจำนวนรอบ

```python
name_list = ['somchai','somying','john']
```

```python
x = 0
for name in name_list:
    print(f'x is {x} and name is {name}')
    x += 1
```

## 2) สร้าง `list` เก็บข้อมูล

```python
name_list = ['somchai','somying','john']
```

```python
x = []
for name in name_list:
    x.append(name)
    print(f'x is {x} and name is {name}')
```

เงื่อนไข `for loop` ใช้เมื่อต้องการสั่งให้การกระทำเกิดกับทุก `object` ใน `object` ที่กำหนด
