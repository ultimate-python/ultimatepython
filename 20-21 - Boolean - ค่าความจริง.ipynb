## 1) Boolean

ข้อมูลประเภทนี้ใช้เก็บค่าความจริงที่มีอยู่ 2 ค่า True และ False ซึ่ง boolean ใช้สำหรับการผูกเงื่อนไขสร้างลำดับการส่งคำสั่งให้ได้ตามความต้องการ

#### ค่าจริง

```python
True
```

#### ค่าเท็จ

```python
False
```

#### พื้นฐานของค่า True False คือ 1 0

#### Object Boolean True มีค่าทางคณิตศาสตร์ 1

```python
True + True
```

#### Object Boolean False มีค่าทางคณิตศาสตร์ 0

```python
True + False
```

## 2) การใช้งาน Boolean 

Boolean จะช่วยในการเขียนโปรแกรมโดยคอยเป็นเหมือน switch ที่คอยบอกให้ทำ หรือไม่ทำคำสั่งต่างๆ ที่เกี่ยวข้องกับ Object ประเภท Boolean ที่อยู่ในโค้ด

หลักการก็คือ เราแบ่งคำสั่งออกเป็นสองชุดที่ไม่เหมือนกัน คือเมื่อได้ค่า Boolean True ให้ทำอย่างนึง ในขณะที่ถ้าได้ False ให้ทำอีกอย่างหนึ่ง

#### กำหนดตัวแปร

```python
a = 3
b = 4
```

#### Equal

```python
a == b
```

#### Not Equal

```python
a != b
```

#### Greater Than

```python
a > b
```

#### Less Than

```python
a < b
```

#### Greater Than or Equal to

```python
a >= b
```

#### Less than or Equal to

```python
a <= b
```

## คำสั่ง in

#### string h เป็น object ใน string hello

```python
'h' in 'hello'
```

#### 1 ที่เป็น integer ที่อยู่ใน list

```python
1 in [1,2,3]
```

#### string h อยู่ใน hello แต่ list ไม่มี h มีแต่ hello

```python
'h' in ['hello']
```

## 4) Chain Comparison Operators

เรียกง่ายๆ ว่าคือการผูกเงื่อนไข ซึ่งมีพื้นฐานมาจากตรรกศาสตร์ คำสั่งที่เราจะใช้บ่อยคือ and และ or

#### and

```python
True and True
```

```python
True and False
```

```python
False and False
```

#### &

```python
True & True
```

#### or

```python
True or True
```

```python
True or False
```

```python
False or False
```

#### |

```python
True | True
```

# ใส่วงเล็บ

```python
c = 10
d = 20
```

```python
(c < d) | (c == d)
```

```python
c < d | c == d
```

#### ข้อสังเกต

```python
d or c
```

```python
d | c
```
