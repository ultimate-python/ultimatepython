## 1)  `break`  `loop`

 `loop`  `break`  `loop`  

```python
while True:
    print('This is print before break')
    break
    print('This is print after break')
```

## 2)  `continue` 

 `loop`  `continue`  `loop` 

```python
name_list = ['somchai','somying','john']
```

```python
for name in name_list:
    if name == 'somchai':
        continue
    print(name)    
```
