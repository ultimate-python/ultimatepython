## 1) คำสั่ง `break` หยุด `loop`

ใช้งานร่วมกับ `loop` เมื่อเจอคำสั่ง `break` จะหยุดการทำงานของ `loop` นั้นๆ ทันที

```python
while True:
    print('This is print before break')
    break
    print('This is print after break')
```

## 2) คำสั่ง `continue` เช็คเงื่อนไขใหม่

ใช้งานร่วมกับ `loop` เมื่อเจอคำสั่ง `continue` จะเป็นคำสั่งให้กลับไปเช็คเงื่อนไขของ `loop` ใหม่ทันที

```python
name_list = ['somchai','somying','john']
```

```python
for name in name_list:
    if name == 'somchai':
        continue
    print(name)    
```

## 3) คำสั่ง `pass` สั่งให้ผ่านไป
