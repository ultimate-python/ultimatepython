## 1) `list` ชุดข้อมูลแบบรายการ

`list` เป็นชุดข้อมูลที่มีการเก็บข้อมูลด้วยตำแหน่งของข้อมูล ซึ่งสิ่งที่ใช้ระบุข้อมูลจะถูกเรียกว่า `index` โดยตำแหน่งของข้อมูลจะเริ่มที่ 0 เสมอ

## 2) สร้าง `list`

สร้าง `list` ที่ไม่บรรจุ `object` ใดๆ ไว้

```python
[]
```

```python
list()
```

## 3) ใส่ข้อมูลใน `list`

ใน `list` สามารถใส่ `object` ประเภทใด กี่ `object` ก็ได้โดยให้แต่ละ `object` แบ่งกันด้วย `,`

```python
['strint',10,10.0]
```

## 4) ดึงข้อมูลด้วยตำแหน่ง Indexing
เราสามารถเรียกข้อมูลใน `list` ด้วยการใช้ Indexing และ Slicing ใน ลักษณะเดียวกันกับ Strings

```python
[1,2,3][0]
```

## 5) ดึงข้อมูลเป็นช่วงแบบ slicing
เราสามารถเรียกข้อมูลใน `list` ด้วยการใช้ Indexing และ Slicing ใน ลักษณะเดียวกันกับ Strings

```python
[1,2,3][1:3]
```

## 6) การรวม `list` ด้วยการ `+`

```python
[1,2,3] + [4]
```

## 7) `list` `methods` คำสั่งที่ใช้บ่อย

```python
list1 = [1,2,3]
```

## 8) `.append()` คำสั่งใส่ `input` เข้าไปใน `list`

```python
list1.append('append me!')
```

```python
list1
```

## 9) `.pop()` คำสั่งเอา `object` ออกจาก `list` โดยการระบุ `index` ใน `input`

`.pop()` ให้ `output` เป็น `object` ที่ถูกนำออกจาก `list`

```python
list1.pop(-1)
```

`.pop()` ยังนำ `object` ออกจาก `list` ด้วย

```python
list1
```
