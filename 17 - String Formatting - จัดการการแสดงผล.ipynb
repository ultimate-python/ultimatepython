## 1) คำสั่ง `print` และ `string`

`string` ถูกสร้างให้มีโครงสร้างที่เหมาะสมกับการใช้แสดงผลด้วยคำสั่ง `print` มากที่สุด

## 2) `string literals` - จัดการการแสดงผล

คือ สัญลักษณ์พิเศษที่เมื่อใส่เข้าไปใน `string` แล้วใช้งานร่วมกับ `function` `print` จะช่วยในการจัดหน้าการแสดงผลให้สวยงาม

### 2.1) `\t` - เว้นช่องว่าง 1 tab

```python
print('hello\tworld')
```

### 2.2) `\n` - ขึ้นบรรทัดใหม่

```python
print('hello\nworld')
```

## 3) `string formatting` - จัดการการแสดงผล

การเขียนแบบ `f-string` โดยการใส่ตัว `f` ไปไว้ด้านหน้าของ `input` สำหรับ `function` `print` เพื่อใช้แสดง `string` ที่ต้องการเข้าไปด้วย

ในลักษณะการเขียนดังนี้ `print(f'string {สิ่งที่อยากแสดง}')`

```python
word = 'hello world'
```

```python
print(f'His name is {word}')
```
