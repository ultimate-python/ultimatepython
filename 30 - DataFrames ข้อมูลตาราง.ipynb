`dataframe` เป็น `object` ที่ถูกสร้างขึ้นด้วย `library` `pandas` ดังนั้นจะใช้งานได้จะต้องเรียก `pandas` มาไว้บน Jupyter Notebook ก่อน

```python
import pandas as pd
```

## 3) สร้าง `dataframe` ผ่าน `pd.DataFrame()`

สร้าง `dataframe` เปล่า

```python
pd.DataFrame()
```

## 4) กำหนด `data` ด้วย `list` `object`

```python
data = [1,2,3,4]
```

## 5) กำหนด `index` ด้วย `list` `object`

```python
index = ['a','b','c','d']
```

## 6) กำหนด `column` ด้วย `list` `object`

```python
column = ['column']
```

# DataFrames

Dataframe เป็นเหมือนตารางที่เกิดจากการรวม series เข้าด้วยกันด้วย index

```python
import pandas as pd
```

```python
table = pd.DataFrame(data=data,index=index,columns=column)
table
```

## 7) การดึงข้อมูลด้วย `column`

ใช้สัญลักษณ์ `[]` เช่นเดียวกับการดึงข้อมูลจากทุก `object` ที่เราได้เรียนไปแล้ว พอดึงแล้วเราจะได้ `object` `series`

```python
table['column']
```

## 8) การดึงข้อมูลด้วย `column` และ `index`

ใช้สัญลักษณ์ `[]` เช่นเดียวกับการดึงข้อมูลจากทุก `object` ที่เราได้เรียนไปแล้ว พอดึงแล้วเราจะได้ `object` `series`

```python
table['column']['a']
```

## 9) การดึงข้อมูลด้วย `index`

ใช้คำสั่ง `.loc[]` โดยใส่ `index` ที่ต้องการดึงไปใน `[]`

```python
table
```

```python
table.loc['a']
```

## 10) การ filter ใน `dataframe`

การ filter ใช้ `list` ของ `boolean` เป็นการกำหนดการโชว์หรือไม่โชว์บรรทัดนั้น

```python
table[[True,True,False,False]]
```

โดยปกติเราจะสร้าง `list` ของ `boolean` ด้วยการเขียนเงื่อนไข

```python
table['column'] < 3
```

```python
table[table['column'] < 3]
```
