## 1) `while loop` ทำไปเรื่อยๆ เมื่อจริง

เงื่อนไข `while loop` ใช้งานร่วมกับ `boolean` เป็นเงื่อนไขที่ทำให้เกิดการกระทำซ้ำไปเรื่อยๆ ในขณะที่รับค่า `True`

## 2) การใช้งาน `while loop`

ใช้คำสั่ง และวิธีการเขียนดังตัวอย่างด้านล่าง

```python
while True:
    print('The statement is true!')
```

## 3) สร้างเงื่อนไข ด้วยการเปรียบเทียบ

`boolean` ที่ใช้ในการกำหนดเงื่อนไขสามารถสร้างได้ด้วยการเปรียบเทียบ

```python
count = 0
```

```python
while count < 5:
    print(count)
    count += 1
```

## 3) การใช้ `else` รองรับค่า `False` 

`boolean` ที่ใช้ในการกำหนดเงื่อนไขสามารถสร้างได้ด้วยการเปรียบเทียบ

```python
count = 0
```

```python
while count < 5:
    print(count)
    count += 1
else:
    print('Now, count is equal to 5!')
```
