## 1) การทำงานกับ cell

ในการจัดการกับเอกสาร .ipynb จะแตกต่างกับการจัดการไฟล์เอกสารปกติอยู่ตรงที่ไฟล์ .ipynb แบ่งเอกสารเป็นส่วนย่อยที่เรียกว่า cell ดังนั้นการกระทำต่อไฟล์จะเน้นไปที่การกระทำต่อ cell ซึ่งครอบคลุมทั้งการจัดหน้าไฟล์เอกสาร และการแก้ไขเนื้อหาภายใน cell

**ซึ่งการทำงานกับ cell จะแบ่งเป็น 2 modes**
    
    1.Command Mode: การทำงานเพื่อจัดหน้าตาของไฟล์
    2.Edit Mode: การแก้ไขเนื้อหาภายใน cell
    
**ข้อสังเกต:** สามารถสังเกตได้จากสีของกรอบของ cell ที่กำลังเลือกอยู่ สีฟ้า แทน Command Mode สีเขียว แทน Edit Mode

## 2) Command Mode

ใน Command Mode การทำงานกับ cell จะเป็นเรื่องการจัดการหน้าไฟล์เอกสาร เพิ่ม ลด ย้าย copy เลื่อน cell และเลือก cell เพื่อทำงานด้วย สัญลักษณ์สำคัญของ Command Mode คือแถบสีฟ้า ด้านหน้า cell

**ตัวอย่างสัญลักษณ์ของ Command Mode**

### คีย์ลัด

#### (i) เลือก cell ที่จะทำงานด้วย

ใช้ arrows `↓` `↑` เพื่อเลื่อนขึ้นลงตามตำแหน่งของ cell ใน notebook ได้

หรือ `คลิก` บนเซลล์เพื่อเลือกเซลล์ที่จะทำงานด้วย

#### (ii) สร้าง cell ใหม่ 

กด `a` เพื่อสร้าง cell ใหม่บนไฟล์ด้านบน (above) cell ที่เราเลือกอยู่

กด `b` เพื่อสร้าง cell ใหม่บนไฟล์ด้านล่าง (below) cell ที่เราเลือกอยู่

#### (iii) เปลี่ยนชนิด cell 

กด `m` เพื่อเปลี่ยน cell ให้เป็น Markdown Cell

กด `y` เพื่อเปลี่ยน cell ให้เป็น Code Cell

#### (iv) ลบ cell

กด `x` เมื่อต้องการลบ cell ที่กำลังเลือกอยู่

กด `z` เมื่อต้องยกเลิกการลบ cell (ครั้งล่าสุด และถัดไปเรื่อยๆ)


#### (v) copy & paste cell

กด `Ctrl` + `c` เพื่อคัดลอก cell ที่กำลังเลือกอยู่

กด `Ctrl` + `v` เพื่อวาง cell ที่คัดลอกไว้ด้านบนของ cell ที่เลือกอยู่

## 3) Edit Mode

Edit Mode เป็นการทำงานแก้ไขเนื้อหาใน cell ซึ่งเนื้อหาใน cell จะต่างกันไประหว่างชนิดของ cell สำหรับ Markdown Cell จะบันทึกข้อมูลที่เป็นคำบรรยาย รายงาน และอื่นๆ ที่เป็นการเขียนโดยใช้ภาษามนุษย์ ในขณะที่ Code Cell จะเป็นการเข้าไปแก้ไขคำสั่งโดยใช้ภาษา Python

**ตัวอย่างสัญลักษณ์ของ Edit Mode**

## 4) เปลี่ยน mode การทำงาน

### คีย์ลัด

#### (i) เปลี่ยน command mode เป็น edit mode

กด `Enter` เปลี่ยนจาก command mode เป็น edit mode

หรือ `ดับเบิลคลิก` ที่ cell เพื่อเปลี่ยนจาก command mode เป็น edit mode

#### (ii) ออกจาก edit mode เป็น command mode

กด `Esc` เพื่อยกเลิก edit mode และ กลับสู่ command mode

หรือ รันชุดคำสั่ง ด้วย `Ctrl` + `Enter` เมื่อรันเสร็จ Jupyter Notebook จะยังเลือก cell เดิม

หรือ รันชุดคำสั่ง ด้วย `Shift` + `Enter` เมื่อรันเสร็จ Jupyter Notebook จะเลื่อนไปยัง cell ด้านล่าง
